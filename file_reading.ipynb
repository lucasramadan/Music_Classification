```python
import pandas as pd
import os
```

```python
data = {'file': [], 'lyrics': []}

for (dirpath, dirnames, filenames) in os.walk('data/'):
    for fn in filenames[1:]:
        with open('data/'+fn, 'r') as f:
            data['lyrics'].append(f.read())
            data['file'].append(fn)
```

```python
df = pd.DataFrame(data)
```
